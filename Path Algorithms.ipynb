```python
# find a path through a graph using recursion
# from Python Patterns - Implementing Graphs http://www.python.org/doc/essays/graphs/

# Input is a graph, a start node and an end node
# Output is a list of lists: each path is contained in a list

graph = {'A': ['B', 'C'],
             'B': ['C', 'D'],
             'C': ['D'],
             'D': ['C'],
             'E': ['F'],
             'F': ['C']}

def find_path(graph, start, end, path=[]):
        path = path + [start]               # path is a list containing a single path
        if start == end:                    # base case
            return path
        if not graph.has_key(start):        # there are no outgoing links from 'start'
            return None  
        for node in graph[start]:
            if node not in path:
                newpath = find_path(graph, node, end, path)
                if newpath: return newpath
        return None
    
print find_path (graph, 'A', 'D')
```

```python
# find the shortest path using recursion
# from Python Patterns - Implementing Graphs http://www.python.org/doc/essays/graphs/

graph = {'A': ['B', 'C'],
             'B': ['C', 'D'],
             'C': ['D'],
             'D': ['C'],
             'E': ['F'],
             'F': ['C']}

def find_shortest_path(graph, start, end, path=[]):
    path = path + [start]
    if start == end:
        return path
    if not graph.has_key(start):
        return None
    shortest = None
    for node in graph[start]:
        if node not in path:
            newpath = find_shortest_path(graph, node, end, path)
            if newpath:
                if not shortest or len(newpath) < len(shortest):
                    shortest = newpath
    return shortest

print find_shortest_path(graph, 'A', 'D')
```

```python
# find a path through a graph using iteration
# use DFS

# Input is a graph, a start node and an end node
# Output is a list of lists: each path is contained in a list

graph = {'A': ['B', 'C'],
             'B': ['C', 'D'],
             'C': ['D'],
             'D': ['C'],
             'E': ['F'],
             'F': ['C']}

graph2 = {'A': ['B', 'C'],
          'B': ['D', 'E'],
          'C': ['F', 'G'],
          'D': ['H'],
          'E': ['H'],
          'F': ['H'],
          'G': ['H'],}

def find_path_iterative (graph, start, end):
    # Find any path from "start" to "end" on a graph.  
    # This approach builds that path as it goes versus requiring a back trace.
    frontier = [[start]]
    current = start
    while frontier:
        path = frontier.pop(0)  # This is a DFS. If we use pop(0) it's BFS
        current = path[-1]
        if graph.has_key(current): 
            for neighbor in graph[current]:
                if neighbor not in path:  # avoids cycles
                    print "current:", current, "neighbor:", neighbor, "path:", path
                    frontier.append(path+[neighbor])
                    print "frontier:", frontier
                    if neighbor == end:
                        return path + [neighbor]
    return None
    
print find_path_iterative (graph2, 'A', 'H')

print "==========================================="

def find_path_iterative_node (graph, start, end):  # Revise this one
        path = [start]
        visited = [start]
        while path[-1] != end:
            current = path[-1]
            if graph.has_key(current): 
                for neighbor in graph[current]:
                    if neighbor not in visited:
                        if neighbor == end:
                            return path + [neighbor]
                        else:
                            path.append(neighbor)
                            visited.append(neighbor)
                            break
        return None
    
print find_path_iterative_node (graph, 'A', 'D')
```

```python
# find all paths through a graph using iteration
# use DFS

# Input is a graph, a start node and an end node
# Output is a list of lists: each path is contained in a list

graph = {'A': ['B', 'C'],
             'B': ['C', 'D'],
             'C': ['D'],
             'D': ['C'],
             'E': ['F'],
             'F': ['C']}

graph2 = {'A': ['B', 'C'],
          'B': ['D', 'E'],
          'C': ['F', 'G'],
          'D': ['H'],
          'E': ['H'],
          'F': ['H'],
          'G': ['H'],}

def find_all_paths_iterative (graph, start, end):
    # Find all path from "start" to "end" on a graph.  
    # This approach builds paths as it goes and does not require a back trace.
    frontier = [[start]]
    solutions = []
    current = start
    while frontier:
        path = frontier.pop()  # This is a DFS. If we use pop(0) it's BFS.
        # And the first path it finds will be the shortest path
        current = path[-1]
        if graph.has_key(current): 
            for neighbor in graph[current]:
                if neighbor not in path:  # avoids cycles
                    print "current:", current, "neighbor:", neighbor, "path:", path
                    frontier.append(path+[neighbor])
                    print "frontier:", frontier
                    if neighbor == end:
                        solutions.append(path + [neighbor])
                        # to make this shortest, just keep a running shortest list
    return solutions
    
print find_all_paths_iterative (graph2, 'A', 'H')

print "==========================================="
```
