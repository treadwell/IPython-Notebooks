```python
import csv as csv
import string

def read_data(filename):
    f = open(filename, 'r')
    reader = csv.reader(f, delimiter = ',')
    paths = [string.replace(row[0].split("),", 1)[1],"\\","/") for row in reader]
    return paths

# "V:\Books\Calibre formats to fix.csv"

filename = 'V:\Books\Calibre formats to fix.csv'
print read_data(filename)
```
