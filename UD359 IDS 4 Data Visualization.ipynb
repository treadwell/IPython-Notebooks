```python
from ggplot import *
from pandas import *

def lineplot():
    path = '/Users/kbrooks/Documents/UD359 IDS/Lesson 4 Data Visualization/'
    file_name = 'hr_year.csv'

    hr_year = pandas.read_csv(path + file_name)
    print ggplot(hr_year, aes('yearID', 'HR')) \
    + geom_point(color = 'red') \
    + geom_line(color = 'red') \
    + ggtitle('Total HRs by Year') \
    + xlab('Year') \
    + ylab('HR')

%pylab inline
print lineplot()
```
